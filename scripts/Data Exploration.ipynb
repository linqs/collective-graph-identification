```python
import pandas as pd
import re
```

```python
# used for extracting feature name

def get_feature(feature):
    feature_data = re.split(r"[:=]", feature)
    return feature_data
    
```

# Loading in the sparse, tab delimited node file

```python
# initialize the pandas dataframe
node_data = pd.DataFrame()

with open('sample-enron.NODE.email.tab') as infile:
  i = 0
  for row in infile:
    if i == 0:
        # Skip non-useful first line
        print("Header: ", row)
    if i == 1:
        # Prepare dataframe column labels
        tokens = row.split()
        features = ["node"] + [get_feature(feature)[1] for feature in tokens]
        node_data = pd.DataFrame(columns=features)
    else:
        # Fill in rows
        tokens = row.split()
        node_data = pd.concat([node_data, pd.DataFrame({'node':[tokens[0]]}.update({}))], ignore_index=True)
        print('i is: ', i)
    i += 1
```
